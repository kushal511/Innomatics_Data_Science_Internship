# Dataset Description

The dataset was released by Aspiring Minds from the Aspiring Mind Employment Outcome 2015 (AMEO). The study is primarily limited  only to students with engineering disciplines. The dataset contains the employment outcomes of engineering graduates as dependent variables (Salary, Job Titles, and Job Locations) along with the standardized scores from three different areas – cognitive skills, technical skills and personality skills. The dataset also contains demographic features. The dataset  contains  around  40 independent variables and 4000 data points. The independent variables are both continuous and categorical in nature. The dataset contains a unique identifier for each candidate.

# Objective

The objective of this analysis is to explore the employment outcomes of engineering graduates using the Aspiring Mind Employment Outcome 2015 (AMEO) dataset. The primary focus is to understand the factors influencing salary, job titles, and job locations for engineering graduates. By conducting exploratory data analysis (EDA) on this dataset, we aim to achieve the following objectives:

(1) Understand Employment Patterns: Investigate the distribution of job titles and job locations among engineering     graduates. Identify the most common job titles and prevalent job locations.

(2) Analyze Salary Distribution: Explore the distribution of salaries among engineering graduates and identify         factors influencing salary variations. Investigate the relationship between demographic features, cognitive         skills, technical skills, personality traits, and salary.

(3) Identify Influential Factors: Determine the factors (both demographic and skill-based) that significantly           impact salary levels and job titles. Investigate correlations between independent variables (such as cognitive     skills, technical skills, personality traits) and dependent variables (salary, job titles).

(4) Evaluate Gender Disparities: Assess if there are any gender disparities in employment outcomes, particularly in     terms of salary levels, job titles, and job locations. Investigate whether gender influences specialization         preferences and salary expectations among engineering graduates.

```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("-"*20, col_name, "-"*20)
        print('Value Counts: ', discrete_data[col_name].value_counts())
        print('\n')
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('\n')
discrete_univariate_analysis(discrete_df)
        
```
